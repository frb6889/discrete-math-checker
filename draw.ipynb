```python
G2 = nx.Graph(
    [(1,5), 
    (5,3), 
    (3,0), 
    (0,2), 
    (2,4),
    (4,1),
    (1,3),
    (2,5),
    (0,4)])
pos2 = nx.circular_layout(G2)
nx.draw(G2,pos2, with_labels=True,node_size=700, node_color='white',edgecolors='black')
```
